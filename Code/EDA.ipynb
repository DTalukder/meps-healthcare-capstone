```python
# 1. Setup and data load

# Install missing libraries
!pip install openpyxl

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set display options
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)

# Load the dataset
file_path = '/content/h243.xlsx'
df = pd.read_excel('h243.xlsx')

print("\nDataset loaded successfully.")
print(f"Dataset Shape: {df.shape}")
display(df.head())



# Legend
variable_legend = {
    'PERWT22F': 'Person Final Weight',
    'FAMWT22F': 'Family Final Weight',
    'FAMWT22C': 'Family CPS Weight',
    'RXOSR22': 'Prescription Cost (Out-of-pocket)',
    'RXPTR22': 'Prescription Cost (Private)',
    'RXOTH22': 'Prescription Cost (Other)',
    'REGION22': 'Region',
    'SEX': 'Sex',
    'RACEV1X': 'Race',
    'AGE22X': 'Age',
}
print("Variable Legend:")
for code, description in variable_legend.items():
    print(f"{code}: {description}")
#



# 2. Initial overview
print("\n--- Dataset Information ---")
df.info()

print("\n--- Descriptive Statistics ---")
display(df.describe())

# 3. Check for missing values
print("\n--- Missing Values Analysis ---")
missing_values = df.isnull().sum()
missing_columns = missing_values[missing_values > 0]
print(missing_columns)
print(f"Total Missing Values: {df.isnull().sum().sum()}")

# 4. Feature mapping
# variable codes to human readable names
feature_names = {
    'PERWT22F': 'Person Final Weight',
    'FAMWT22F': 'Family Final Weight',
    'FAMWT22C': 'Family CPS Weight',
    'RXOSR22': 'Prescription Cost (Out-of-pocket)',
    'RXPTR22': 'Prescription Cost (Private)',
    'RXOTH22': 'Prescription Cost (Other)',
    'REGION22': 'Region',
    'SEX': 'Sex',
    'RACEV1X': 'Race',
    'AGE22X': 'Age'
}


df_renamed = df.rename(columns=feature_names)
print("\n--- Sample of Renamed Features ---")
print(df_renamed.columns[:15])

# 5. Correlation Analysis (Numerical)
print("\n--- Correlation Analysis ---")
numerical_df = df.select_dtypes(include=[np.number])
correlation_matrix = numerical_df.corr()

# Heatmap of top correlations (top 15 feature pairs)
top_corr = correlation_matrix.abs().unstack().sort_values(ascending=False).drop_duplicates()
top_features = list(set([idx[0] for idx in top_corr[:15].index] + [idx[1] for idx in top_corr[:15].index]))

plt.figure(figsize=(12, 10))
sns.heatmap(numerical_df[top_features].corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Top Correlations Among Numerical Features")
plt.show()

# 6. Outlier Detection
print("\n--- Outlier Detection ---")
cols_to_plot = ['PERWT22F', 'FAMWT22F', 'FAMWT22C']
for col in cols_to_plot:
    if col in df.columns:
        plt.figure(figsize=(8, 4))
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot of {col}')
        plt.show()

# 7. Distribution analysis of expense variables and skewness quantification
print("\n--- Expense Variables Distribution & Skewness ---")
expense_cols = ['RXOSR22', 'RXPTR22', 'RXOTH22']
for col in expense_cols:
    if col in df.columns:
        plt.figure(figsize=(8, 4))
        sns.histplot(df[col], bins=50, kde=True)
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()
        skewness = df[col].skew()
        print(f"Skewness of {col}: {skewness:.2f}")

# 8. Categorical variable distributions
print("\n--- Categorical Variables Distribution ---")
categorical_vars = ['REGION22', 'SEX', 'RACEV1X']
for var in categorical_vars:
    if var in df.columns:
        plt.figure(figsize=(8, 5))
        sns.countplot(data=df, x=var, order=df[var].value_counts().index)
        plt.title(f'Distribution of {var}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

# 9. Advanced analysis: Demographic impacts on healthcare spending

# 9a. Create age bins
if 'AGE22X' in df.columns:
    bins = [0, 18, 35, 50, 65, 100]
    labels = ['<18', '18-35', '35-50', '50-65', '65+']
    df['AGE_BIN'] = pd.cut(df['AGE22X'], bins=bins, labels=labels, right=False)
    plt.figure(figsize=(8, 5))
    sns.countplot(x='AGE_BIN', data=df, order=labels)
    plt.title("Age Distribution (Binned)")
    plt.xlabel("Age Groups")
    plt.ylabel("Count")
    plt.show()

# 9b. Analyze healthcare expenditures by demographics
if 'TOTEXP22' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='REGION22', y='TOTEXP22', data=df)
    plt.title("Total Healthcare Expenditure by Region")
    plt.xlabel("Region")
    plt.ylabel("Total Expenditure")
    plt.show()

    if 'AGE_BIN' in df.columns:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x='AGE_BIN', y='TOTEXP22', data=df, order=labels)
        plt.title("Total Healthcare Expenditure by Age Group")
        plt.xlabel("Age Groups")
        plt.ylabel("Total Expenditure")
        plt.show()
else:
    print("\nNote: 'TOTEXP22' column not found. Consider analyzing other cost-related variables for demographic impacts.")

# 10. (note to self) save the cleaned data
df.to_csv('meps_cleaned.csv', index=False)
print("\nCleaned data saved as 'meps_cleaned.csv'.")

# 11. Summary of insights
print("\n--- SUMMARY OF FINDINGS ---")
print("1. Data Shape:", df.shape)
if df.isnull().sum().sum() == 0:
    print("2. No missing values detected.")
else:
    print("2. Missing values detected in some columns.")
print("3. Strong correlations identified (e.g., between 'PERWT22F' and 'FAMWT22F').")
print("4. Outliers found in weight-related columns (PERWT22F, FAMWT22F, FAMWT22C).")
print("5. Expense variables such as 'RXOSR22', 'RXPTR22', and 'RXOTH22' are heavily skewed.")
print("6. Demographic distributions (Region, Sex, Race, Age) provide a strong basis for further segmentation and modeling.")
```
